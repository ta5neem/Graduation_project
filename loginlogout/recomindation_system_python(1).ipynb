```python
import json
app = FastAPI()



class UserRequestIn(BaseModel):
    text: str
    id : int
    image : str   
    arr : list 
        
    

    
@app.get("/sugg_recm/{name1}")    
def sugg_recm(name1):
    
    
    db_connection_str = 'mysql+pymysql://root:@127.0.0.1/laravel'
    db_connection = create_engine(db_connection_str)

  




    mydb = mysql.connector.connect(
      host="127.0.0.1",
      user="root",
      password="",
      database="laravel"
    )

    mycursor = mydb.cursor()
    id = "select id from products where name='%s'" % (name1)
    mycursor.execute(id)
    id = mycursor.fetchall()
    
    sql = "select * from order_details where product_id=  (%s )" % (id[0][0])
    mycursor.execute(sql)
    myresult = mycursor.fetchall()
    i=0
    j=0

    myresult2=[]
    myresult3=[]
    myresult4=[]



    for item in myresult :
        mycursor = mydb.cursor()
        sql = "select * from orders where id=(%s)" % (item[6])
        mycursor.execute(sql)


        myresult2.append(mycursor.fetchall())

        i=i+1
        j=0
        for item2 in myresult2 :
            mycursor = mydb.cursor()
            sql = "select * from order_details where order_id=(%s)" % (myresult2[j][0][0])
            mycursor.execute(sql)

            myresult3.append(mycursor.fetchall())
            j=j+1
            k=0
            for item3 in myresult3 :
                for l in range( len(myresult3[k])):
                    mycursor = mydb.cursor()
                    sql = "select * from products where id=(%s)" % (myresult3[k][l][5])
                    mycursor.execute(sql)

                    myresult4.append(list(mycursor.fetchall()))
                k=k+1

    df = pd.read_sql(sql, con=db_connection)



    df = pd.DataFrame(myresult4)

    df2 = pd.DataFrame(df[0].tolist(), index=df.index)


    df2.drop_duplicates( keep = 'first', inplace = True)


        




    df2.columns = ['id','name','sell_price','total_q','description','color','size','image','brand_id','category_id','vote_count','vote_average','popularity','created_at','updated_at']


    #return df2
    
    commentList = []
    #movies_df = pd.read_csv(r"D:\New folder\Downloads\product.csv") 
    
    
    df2 =  df2.drop(['sell_price', 'total_q', 'size', 'brand_id'], axis = 1)
    #movies_df.head()


    V = df2[ df2['vote_count'].notnull()]['vote_count'].astype('float')
    #vote_average
    R =  df2[ df2['vote_average'].notnull()]['vote_average'].astype('float')
    #vote_average_mean
    C = R.mean()
    #minimum votes required to get in list
    M = V.quantile(0.1)


    df = pd.DataFrame()
    df =  df2[( df2['vote_count'] >= M) & ( df2['vote_average'].notnull())][['name','color','vote_count','vote_average','popularity','description','category_id']]



    df['Weighted_average'] = ((R*V) + (C*M))/(V+M)
    recm_movies = df.sort_values('Weighted_average', ascending=False).head(500)
    #recm_movies.head()



    popular = pd.DataFrame()
    popular =  df2.copy()
    popular['popularity'] = recm_movies[recm_movies['popularity'].notnull()]['popularity'].astype('float')
    popular = popular.sort_values('popularity',ascending = False)



    s = df.apply(lambda x: pd.Series(x['category_id']),axis=1).stack().reset_index(level=1, drop=True)
    s.name = 'category_id'
    gen_movies = recm_movies.drop('category_id', axis=1).join(s)
    df['Weighted_Pubularity']=df['Weighted_average']+df['popularity']
    recm_movies = df.sort_values('Weighted_Pubularity', ascending=False).head(500)




    #recm_movies.head()
    df_w = gen_movies[ (gen_movies['category_id'] == 2) & (gen_movies['vote_count'] >= M)]
    df_w.sort_values('Weighted_average', ascending = False).head()
    df_w = df_w.sort_values('Weighted_average', ascending = False)
    plt.figure(figsize=(12,6))
    #axis1=sns.barplot(x=df_w['Weighted_average'].head(10), y=df_w['name'].head(10), data=df_w)





    cont_recm = recm_movies.copy()
   # cont_recm.head()



    tfv = TfidfVectorizer(min_df=3, max_features=None, strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
                ngram_range=(1, 3), stop_words = 'english')
    cont_recm['color'] = cont_recm['color'].fillna('')


    tfv_matrix = tfv.fit_transform(cont_recm['color'])
    #Finding Cosine_similarity
    cos_sim = linear_kernel(tfv_matrix, tfv_matrix)
    cont_recm = cont_recm.reset_index()
    indices = pd.Series(cont_recm.index, index=cont_recm['name'])
    #ser = pd.Series(data=d, index=['a', 'b', 'c'])


    # Get the pairwsie similarity scores 
    sim_scores = list(enumerate(cos_sim[indices[name1] ]))
    # Sort the movies 
    sim_scores = sorted(sim_scores, key=lambda x: x[1],reverse=True)
    id = 5
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores] 
    result =df2.iloc[movie_indices]['id']
    #commentList.append(result.iloc[0])
    id=result.iloc[0]
    
#     recm_movies = recm_movies[(recm_movies['name']!= name1)]
#     recm_movies  = recm_movies[ (recm_movies['category_id'] != 2)]
#     return recm_movies.iloc[movie_indices]['name']

#     d=df2.loc[name1,"category_id"]
#     df2  = df2[ (df2['category_id'] != d and df2['name'])]
    
    for i in range(5):   
        commentList.append(int(result.iloc[i]))
    
#     res = json.dumps(commentList)
#     return res
    return {
      "list":  commentList
    }


def delete(name):
    file = pathlib.Path(r"D:\New folder\Downloads\product(2).csv")
    file.unlink()
    
  
    
```
